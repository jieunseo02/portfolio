```python
# BMI

import pandas as pd
import matplotlib.pyplot as plt

file_path = "C:/Users/강준희/Desktop/한림대/데이터분석 대회/데이터/BMI.xlsx"
df = pd.read_excel(file_path)

bmi = 'BMI'

mean = round(df[bmi].mean(),3)
variance = round(df[bmi].var(),3)
std_deviation = round(df[bmi].std(),3)

print(f'평균: {mean}')
print(f'분산: {variance}')
print(f'표준편차: {std_deviation}')

plt.figure(figsize=(10,6))
plt.hist(df[bmi],bins=30,edgecolor = 'black')
plt.title('Histogram of '+ bmi)
plt.xlabel(bmi)
plt.grid(axis='y',alpha=0.75)
plt.show()
```

```python
# BMI와 LDLC 상관관계

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

bmi_file_path = "C:/Users/강준희/Desktop/한림대/데이터분석 대회/데이터/BMI.xlsx"
df_bmi = pd.read_excel(bmi_file_path)

ldlc_file_path = "C:/Users/강준희/Desktop/한림대/데이터분석 대회/데이터/LDLC.xlsx"
df_ldlc = pd.read_excel(ldlc_file_path)

df = pd.concat([df_bmi['BMI'], df_ldlc['LDLC']], axis=1)
df.columns = ['BMI', 'LDLC']

correlation = df.corr()
print(f'상관 계수 행렬:\n{correlation}')

pearson_corr = correlation.loc['BMI', 'LDLC']
print(f'BMI와 LDLC의 Pearson 상관 계수: {pearson_corr:.3f}')

plt.figure(figsize=(10, 6))
sns.scatterplot(x='BMI', y='LDLC', data=df)
plt.title('Scatter Plot of BMI and LDLC')
plt.xlabel('BMI')
plt.ylabel('LDLC')
plt.grid(True)
plt.show()

plt.figure(figsize=(6, 4))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of BMI and LDLC')
plt.show()
```

```python
# 로지스틱 회귀분석

import pandas as pd
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

bmi_file = "C:/Users/강준희/Desktop/한림대/데이터분석 대회/데이터/BMI.xlsx"
ldlc_file = "C:/Users/강준희/Desktop/한림대/데이터분석 대회/데이터/LDLC.xlsx"
hdlc_file = "C:/Users/강준희/Desktop/한림대/데이터분석 대회/데이터/HDLC.xlsx"
tg_file = "C:/Users/강준희/Desktop/한림대/데이터분석 대회/데이터/TG.xlsx"
chol_file = "C:/Users/강준희/Desktop/한림대/데이터분석 대회/데이터/Total cholesterol.xlsx"

df_bmi = pd.read_excel(bmi_file)
df_ldlc = pd.read_excel(ldlc_file)
df_hdlc = pd.read_excel(hdlc_file)
df_tg = pd.read_excel(tg_file)
df_chol = pd.read_excel(chol_file)

df = pd.concat([df_bmi, df_ldlc, df_hdlc, df_tg, df_chol], axis=1)

print(df.isna().sum())

numeric_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in numeric_cols:
    df[col].fillna(df[col].mean(), inplace=True)

df.replace([float('inf'), float('-inf')], float('nan'), inplace=True)

print(df.isna().sum())

df['Obesity'] = (df['BMI'] >= 25).astype(int)

df['Abnormal_Lipid'] = ((df['LDLC'] >= 130) | (df['HDLC'] < 40) | (df['TotalC'] >= 240)).astype(int)

X = df[['Obesity', 'LDLC', 'HDLC', 'TG', 'TotalC']]

y = df['Abnormal_Lipid']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train_scaled = sm.add_constant(X_train_scaled)
X_test_scaled = sm.add_constant(X_test_scaled)

logit_model = sm.Logit(y_train, X_train_scaled)
result = logit_model.fit()

print(result.summary())

y_pred = result.predict(X_test_scaled)
y_pred_class = (y_pred >= 0.5).astype(int)

accuracy = (y_pred_class == y_test).mean()
print(f'모델 정확도: {accuracy:.3f}')
```
